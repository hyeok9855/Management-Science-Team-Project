```python
import re
import json

# https://github.com/southkorea/seoul-maps
with open('seoul_municipalities_geo_simple.json', 'r', encoding='utf-8') as f:
    geo = json.load(f)
```

```python
geo
```

```python
seoul_geo = {}
seoul_geo["type"] = "FeatureCollection"
seoul_geo["features"] = []
for dic in geo["features"]:
    new_dic = {}
    new_dic['type'] = 'Feature'
    new_dic['id'] = dic['properties']['SIG_KOR_NM']
    new_dic['properties'] = dic['properties']
    new_dic['geometry'] = dic['geometry']
    seoul_geo["features"].append(new_dic)
```

```python
len(seoul_geo["features"])
```

```python
with open('Seoul_map.json', 'w') as f:
    json.dump(seoul_geo, f)
```
